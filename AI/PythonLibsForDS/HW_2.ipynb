```python
t = np.linspace(0, 10, 51)
```

```python
f = np.cos(t)
```

```python
plt.plot(t, f, color="green")
plt.title("f(t) Graph")
plt.xlabel("x values")
plt.ylabel("y values")
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

```python
x = np.linspace(-3, 3, 51)
```

```python
y1 = np.power(x, 2)
```

```python
y2 = np.add(np.multiply(x, 2), 0.5)
```

```python
y3 = np.add(np.multiply(x, -3), -1.5)
```

```python
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title("y1 Graph")
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title("y2 Graph")

ax3.plot(x, y3)
ax3.set_title("y3 Graph")

ax4.plot(x, y4)
ax4.set_title("y4 Graph")
```
